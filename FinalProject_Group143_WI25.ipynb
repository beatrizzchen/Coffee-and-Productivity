## Background and Prior Work

Caffeine is one of the most widely consumed psychoactive substances in the world, with coffee, tea, and energy drinks serving as primary sources for students seeking to enhance productivity while studying. The effectiveness of caffeine as a cognitive enhancer has long been debated, with studies highlighting both benefits and potential drawbacks. Given its widespread use among students, understanding the relationship between caffeine intake and productivity while studying is essential.

A meta-analysis by Liu et al. (2024) examined the relationship between caffeine intake and anxiety levels, concluding that caffeine consumption, especially at high doses exceeding 400 mg per day, significantly increases the risk of anxiety<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). Since anxiety can negatively impact efficiency, this raises questions about caffeine’s overall effect on student productivity. While caffeine may provide temporary alertness, it may also contribute to heightened stress, potentially undermining sustained academic performance. According to a series by the American Medical Association, 85% of adults consume approximately 135 mg of caffeine per day in the US, which translates to 12 ounces of coffee <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). The amount of caffeine highly depends on the type of brewed coffee, varying from type of bean, grind and drink type. 

Another important aspect of caffeine consumption is its effect on sleep and cognition. Progress in Brain Research, chapter “Effects of caffeine on sleep and cognition” suggests that caffeine effectively counteracts sleep deprivation-related cognitive impairment by stimulating the central nervous system. However, it also disrupts sleep quality, leading to increased daytime sleepiness, which can, in turn, hinder productivity<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). This cyclical pattern suggests that students relying on caffeine for productivity may ultimately experience diminishing returns due to its interference with natural sleep rhythms.

Additionally, a blog post from Psychology Today highlights the paradox of caffeine consumption and productivity. While caffeine can temporarily enhance alertness, it disrupts adenosine receptor function, which can lead to energy depletion and increased stress over time. The article argues that excessive caffeine intake may compromise long-term productivity by creating dependence and promoting poor sleep habits<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). This suggests that while caffeine may provide short-term benefits, its long-term effects on productivity are potentially detrimental.

While research has examined caffeine’s effects on mental health and cognitive function separately, there remains a need to analyze these factors within the context of student productivity. This study aims to address this gap by exploring whether caffeine consumption improves or hinders studying efficiency in the 18-23 age group.

```python
#We're checking to see the different coffee drinks they've collected in order to start standardizing them into numerical data
df2_filtered['CoffeeType'].unique() 
```

```python
# Renaming CoffeeType variables to make it easier to read and type
def rename(coffeetype):
    if coffeetype == 'Double Espresso (Doppio)':
        return 'Double Espresso'  # Rename this value
    elif coffeetype == 'Espresso (Short Black)':
        return 'Espresso'  # Rename this value
    else:
        return coffeetype  # Keep all other values unchanged

# Apply the rename function to the 'CoffeeType' column
df2_filtered['CoffeeType'] = df2_filtered['CoffeeType'].apply(rename)

# Display to see if the function worked
df2_filtered['CoffeeType'].unique()
```

Now that we filtered the columns we'll need, we're going to have to standardize them now. First, we'll be turning the cups of coffee into fluid ounces. However, we don't know the exact fluid ounces present in each coffee cup. Not only that, but there are several types of coffees individuals noted, affecting how much caffeine (mg) and coffee (fl oz.) they've actually consumed. A caffe latte may be consumed in a standard 12 oz. cup while Turkish coffee may be drank from a traditional finjān cup that's only 2-3 ounces.

We're going to go on a whim and assume that the Caffè latte's and Americanos consumed follow the standard coffee size, which is a double shoot of espresso (2 ounces) in a 12 ounce cup, with caffeine levels to be estimated to be 128 mg. Turkish coffee is typically served in 2.5 ounce cups containing 62.5 mg of caffeine. Similarly, an Espresso (short black) is a single shot of espresso (typically 30ml/1oz) with 63 mg of caffeine and a doppio is double that. While Nescafe is a brand and doesn't specify which type of roast and specific product inidividuals used, we're going to assume it's a 'Single-Serve Product' made at home listed by Nescafe. If the individuals follow the specified directions, a tsp of the concentrate followed by 6 oz. of water, serves 30-40 mg of caffeine. Typically, this serving is doubled in order to create a standard 12 oz. cup of coffee, so we'll be doubling this particular coffee type. As you can tell, we have a bunch of coffee variations we need to convert and standardize.

We understand that this is all based on assumptions of what we already know without actually having the correct background information. We accept that this may be inaccurate and may skew certain results. 

```python
#We're going to start by converting the 'CoffeeCupsPerDay' and 'CoffeeType' into fluid ounces and milligrams of caffeine. 

coffee_conversion = {
    'Caffè latte': {'fluid_oz_per_cup': 12, 'caffeine_mg_per_cup': 128}, 
    'American Coffee': {'fluid_oz_per_cup': 12, 'caffeine_mg_per_cup': 128},
    'Nescafe': {'fluid_oz_per_cup': 12, 'caffeine_mg_per_cup': 70},
    'Turkish': {'fluid_oz_per_cup': 2.5, 'caffeine_mg_per_cup': 62.5},
    'Espresso': {'fluid_oz_per_cup': 1, 'caffeine_mg_per_cup': 64},
    'Double Espresso' : {'fluid_oz_per_cup': 2, 'caffeine_mg_per_cup': 128}, 
    'Cappuccino' : {'fluid_oz_per_cup': 6, 'caffeine_mg_per_cup': 64 }
}

#Caffeine per cup function for a column 
def get_caffeine_per_cup(row, coffee_conversion):
    coffee_type = row['CoffeeType']
    return coffee_conversion.get(coffee_type, {'caffeine_mg_per_cup': 0})['caffeine_mg_per_cup']

df2_filtered['TotalFluidOz'] = df2_filtered.apply(lambda row: row['CoffeeCupsPerDay'] * coffee_conversion[row['CoffeeType']]['fluid_oz_per_cup'], axis=1)
df2_filtered['TotalCaffeineMg'] = df2_filtered.apply(lambda row: row['CoffeeCupsPerDay'] * coffee_conversion[row['CoffeeType']]['caffeine_mg_per_cup'], axis=1)
df2_filtered['CaffeinePerCup'] = df2_filtered.apply(lambda row: get_caffeine_per_cup(row, coffee_conversion), axis=1)


# Reorder columns to move TotalFluidOz and TotalCaffeineMg to the front
columns_order = ['CodingHours', 'TotalFluidOz', 'CoffeeCupsPerDay', 'CaffeinePerCup', 'TotalCaffeineMg', 'CoffeeType', 'CoffeeSolveBugs', 'AgeRange']
df2_filtered = df2_filtered[columns_order]

df2_filtered.head()
```

Before concluding that none of the variables have a correlation with productivity scores, we tested whether combinations of variables might have an effect. Here are some key takeaways we have discovered from the table:
- Some interaction terms between variables have smaller effects, indicating that the combined influence of two factors is generally weaker than individual factors 
- The fact that most interaction terms are small suggests that combined effects might not be very strong, but they still provide insight into how different variables work together to affect productivity.

Now that we've visualized most of our data, which seems to show little to no correlation, let's code an actual linear regression graph and see if there's any statistical relationship between caffeine and productivity (coding hours). 

```python
outcome_1, predictors_1 = patsy.dmatrices('CodingHours ~ TotalCaffeineMg', data=df2_filtered)
model_1 = sm.OLS(outcome_1, predictors_1)
results_1 = model_1.fit()
print(results_1.summary())
```

We're going to first look at the R-squared results. The OLS Regression results only showcase a 4.4% of the variance in CodingHours is explained by the amount of milligrams in TotalCaffeineMg. A low R-Squared results suggest that TotalCaffeineMg alone is **not a strong predictor of productivity**. 

The Coefficient results for TotalCaffieneMg only displays 0.0027, meaning that for every 1mg increase of caffeine, CodingHours only increases by 0.0027 of an hour, which is approximately 16 minutes. Yet again, the coefficient is very small, indicating a **weak relationship.** 

Lastly, the P-value for TotalCaffeineMg is 0.112, which is a value greater than 0.05. This means that the relationship between caffeine and coding hours is **not stastically significant.**

Let's delve deeper and look at the different results from each coffee type and see if they differ, as some coffee's are stronger than others.

# Ethics & Privacy

One primary ethical concern is participant privacy, as data on caffeine consumption, study habits, sleep patterns, and anxiety levels can be sensitive. If self-reported survey data is used, responses should be anonymized, and participants should provide informed consent detailing how their data will be collected, stored, and used. When using secondary datasets, we must ensure they comply with relevant privacy policies and terms of use to avoid ethical violations.

Bias in data collection is another major concern. If participants are recruited from a single university, the dataset may not be representative of the broader student population. Additionally, self-reported data can introduce recall bias, as participants may misestimate their intake or study efficiency. To address this, we should use datasets where objective measures such as wearable sleep trackers or cognitive performance tests were used.

Post-analysis, it is important to communicate findings responsibly, avoiding deterministic conclusions. Caffeine's effects on productivity are influenced by multiple factors, so conclusions should reflect this. Additionally, researchers must consider how their findings might be misinterpreted, particularly if the study is used to promote excessive caffeine consumption or reinforce unhealthy study habits.

To address these concerns, we will prioritize transparency in data handling and apply statistical techniques to mitigate bias. When using publicly available datasets, such as those on Kaggle, we will assess their sources and limitations to ensure ethical use.
